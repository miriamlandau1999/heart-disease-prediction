```python
#Observe distribution
df.hist(bins=5)
```

```python
#Observe Correlation
df.corr().style.background_gradient()
```

```python
#print highly correlated features
corr = df.corr()
high_corr = corr[(corr.abs() > 0.5) & (corr.abs() < 1.0)]
print(high_corr.dropna(how='all').dropna(axis=1, how='all'))
```

```python
#Observe correlation with target variable 
df.corr()['target'].sort_values(ascending=False)
```

## key Takeaways
- distribution: 
- **cp, thalach, and slope** have some positive correlation with the target variable
- **thal, ca, exang, oldpeak** have some negative correlation with target variable
- **oldpeak  and slope** are highly correlated with each other at .5
